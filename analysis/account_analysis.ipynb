```python
# Read in balance data
import json

FN = "../account_state_fetch/all_accounts.log"
f = open(FN, 'r') 
account_balances = []
for line in f:
    d = json.loads(line)
    account_balances.append(d)            
f.close()
```

```python
# what are the unique symbols
unique_symbols = set()
for accnt in account_balances:
    for _addr, balances in accnt.items():      
        for symbol in balances['borrows'].keys():
            unique_symbols.add(symbol)
        for symbol in balances['deposits'].keys():
            unique_symbols.add(symbol)
print(unique_symbols)
```

```python
def symbolsForAccount(accnt):
    symbols = set()
    for _addr, balances in accnt.items():      
        for s, b in balances['borrows'].items():
            symbols.add(s)
        for s, b in balances['deposits'].items():
            symbols.add(s)
    return symbols
```

```python
#
# Look at net (supplied - borrowed) for each account
#
import pandas as pd
results = []
for accnt in account_balances:
    for _addr, balances in accnt.items():   
        net = {}
        symbols = symbolsForAccount(accnt)
        for s in symbols:
            net[s] = balances['deposits'].get(s, 0) - balances['borrows'].get(s, 0)
    net['accnt'] = _addr
    results.append(net)
df = pd.DataFrame(results)
df= df.fillna(0)
```

### Fetch current prices from coingecko

```python
ENABLE_LIVE_FETCH = False
```

```python
def fetchPriceFromTokenAddress(token_address, vs_currency = 'usd'):  
        url = "https://api.coingecko.com/api/v3/simple/token_price/ethereum"
        
        params = {"vs_currencies" : vs_currency,
                  "contract_addresses": [token_address]}
        r = requests.get(url, params = params)
        x= json.loads(r.content)
        return float(x[token_address.lower()]['usd'])
       
```

```python
fetchPriceFromTokenAddress("0xC02aaA39b223FE8D0A0e5C4F27eAD9083C756Cc2")
```

```python
if ENABLE_LIVE_FETCH:   #can skip cg fetch
    retries = []
    for token_addr in token_info.keys():
        try:
            price = fetchPriceFromTokenAddress(token_addr)
            print(token_addr, token_addr_to_symbol[token_addr], price)
            time.sleep(6)  #public api has a rate limit of 10-30 calls per minute
        except:
            time.sleep(20)
            print(f"Error fetching price for {token_addr} {token_addr_to_symbol[token_addr]}.")
            price = 0
            retries.append(token_addr)      
        prices[token_addr_to_symbol[token_addr]]= price
    for token_addr in retries:
        try:
            price = fetchPriceFromTokenAddress(token_addr)
            print(token_addr, price)
            time.sleep(6)  #public api has a rate limit of 10-30 calls per minute
        except:
            time.sleep(20)
            print(f"second Error fetching price for {token_addr} {token_addr_to_symbol[token_addr]}. setting price to zero")
            price = 0   
        prices[token_addr_to_symbol[token_addr]]= price
        
    with open("current_prices.json", "w") as write_file:
        _x = json.dump(prices, write_file)
```

### Estimate bad debt per account in USD

```python
column_view = ['accnt','usd_value', 'bad_debt', 'WETH', 'USDC', 'USDT','stETH', 'wstETH', 'WBTC', 'cbETH', 'DAI']
```

```python
def f(row):
    value = 0
    for s in unique_symbols:
        value += row[s] * prices[s]
    return value
df['usd_value'] = df.apply(f, axis = 1)

def f2(row):
    if row['usd_value'] < 0:
        return abs(row['usd_value'])
    return 0
df['bad_debt'] = df.apply(f2, axis = 1)

```

```python
df = df.sort_values(by=['usd_value'], ascending = True)
df[column_view].head(50)
```
